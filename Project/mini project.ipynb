```python
class character:
    
    def __init__(self,name):
        self.name=name
        self.__life=3
        self.__score=0
    def kicked (self):
        self.__score+=10
    def Punched (self):
        self.__score+=5
    def Stabbed (self):
        self.__life-=1
    def displaylife(self):
        return f" The {self.name} has {self.__life} life"
    def displayscore(self):
        return f" The {self.name} has {self.__score} score"
obj=character("Mario")
(obj.kicked())
(obj.Punched())
(obj.Stabbed())
print(obj.displaylife())
print(obj.displayscore())
```

```python

class User:
    num_of_users = 0

    def __init__(self, name, gender, salary):
        self.name = name
        self.gender = gender
        self.salary = salary
        User.num_of_users += 1
        self.unique_account_no = User.num_of_users

    def showdetails(self):
        print("Name:", self.name)
        print("Gender:", self.gender)
        print("Salary:", self.salary)
        print("Unique Account No:", self.unique_account_no)


class Bank:
    __balance = 0

    def __init__(self, name, gender, salary):
        self.user = User(name, gender, salary)

    def deposit(self, amount):
        Bank.__balance += amount

    def withdraw(self, amount):
        if amount > Bank.__balance:
            print("Insufficient Balance")
            self.viewbalance()
        elif amount < 100:
            print("You cannot withdraw less than 100")
            self.viewbalance()
        else:
            Bank.__balance -= amount
            print("Thank you for visiting")
            self.viewbalance()

    def viewbalance(self):
        self.user.showdetails()
        print("Balance:", Bank.__balance)

    def transfer(self, amt, user):
        if amt > Bank.__balance:
            print("Insufficient Balance")
            self.viewbalance()
        elif amt < 1:
            print("You cannot transfer less than 1")
            self.viewbalance()
        else:
            Bank.__balance -= amt
            user.bank.deposit(amt)
            print(amt, "transferred successfully")
            self.viewbalance()

# create two users and two bank accounts
user1 = User("John", "Male", 50000)
user2 = User("Jane", "Female", 60000)

bank1 = Bank("Bank1", "Male", 100000)
bank2 = Bank("Bank2", "Female", 200000)

# deposit some amount in the bank accounts
bank1.deposit(5000)
bank2.deposit(10000)

# view the account details and balance
bank1.viewbalance()  # output: Name: Bank1, Gender: Male, Salary: 100000, Unique Account No: 1, Balance: 5000
bank2.viewbalance()  # output: Name: Bank2, Gender: Female, Salary: 200000, Unique Account No: 2, Balance: 10000

# withdraw some amount from the bank accounts
bank1.withdraw(2000)  # output: Thank you for visiting, Name: Bank1, Gender: Male, Salary: 100000, Unique Account No: 1, Balance: 3000
bank2.withdraw(500)  # output: You cannot withdraw less than 100, Name: Bank2, Gender: Female, Salary: 200000, Unique Account No: 2, Balance: 10000
bank2.withdraw(15000)  # output: Insufficient Balance, Name: Bank2, Gender: Female, Salary: 200000, Unique Account No: 2, Balance: 10000

# transfer some amount from one bank account to another
bank1.transfer(1000, user2.bank)  # output: 1000 transferred successfully, Name: Bank1, Gender: Male, Salary: 100000, Unique Account No: 1, Balance: 2000
                                 #         Name: Jane, Gender: Female, Salary: 60000, Unique Account









```
